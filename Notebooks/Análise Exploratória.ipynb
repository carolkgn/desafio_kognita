## Balanceamento da base

```python
# Distribuição do default
df['default'].value_counts()
```

```python
fig, ax = plt.subplots()
sns.countplot(x='default', data = df, ax=ax, palette = 'magma')
ax.set_title('Distribuição de default na base')
plt.plot()
```

```python
# Boxplot da relação default x valor_quitado
sns.set_theme(style='whitegrid')
sns.boxplot(x='default', y='valor_quitado', data=df)
sns.stripplot(x='default', y='valor_quitado', data=df, palette='crest')
```

```python
# Boxplot da relação default x valor_por_vencer
sns.set_theme(style='whitegrid')
sns.boxplot(x='default', y='valor_por_vencer', data=df)
sns.stripplot(x='default', y='valor_por_vencer', data=df, palette='viridis')
```

```python
# Boxplot da relação default x valor_vencido
sns.set_theme(style='whitegrid')
sns.boxplot(x='default', y='valor_vencido', data=df)
sns.stripplot(x='default', y='valor_vencido', data=df, palette='cubehelix')
```

```python
# Relação entre valor_quitado x valor_por_vencer
sns.set_theme(style='whitegrid')
sns.relplot(x='valor_quitado', y='valor_por_vencer', hue='default', data=df, palette='cubehelix')
```

```python
# Boxplot da relação quant_acao_judicial x acao_judicial_valor
sns.set_theme(style='whitegrid')
sns.relplot(x='acao_judicial_valor', y='quant_acao_judicial', hue='default', data=df, palette='rocket')
```
