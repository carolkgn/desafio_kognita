# Distribuições de variáveis com maiores correlações entre si

## Preparação

### Default 

```python
default = df['default']
default = default.value_counts()
default 
```

### Default 3 meses (dados brutos)

```python
default3 = df['default_3months']
df_default3 = default3.value_counts().to_frame()
df_default3.reset_index()
```

```python
df_transposto = df_default3.T
df_transposto
```

```python
print('default NaN: \n', default.isna().value_counts())
print('********************')
print('default 3 meses NaN: \n', default3.isna().value_counts())
```

### Default 3 meses sem contagem zero

```python
default3_maior_que_zero = df.query('default_3months > 0')
```

```python
default3_maior_que_zero = default3_maior_que_zero['default_3months']
```

```python
default3_maior_que_zero.isna()
```

```python
print('default 3 meses (maior que zero) NaN: \n', default3_maior_que_zero.isna().value_counts())
```

```python
default3_maior_que_zero = default3_maior_que_zero.value_counts()
```

```python
df_default3_maior_que_zero = default3_maior_que_zero.to_frame()
```

```python
df_default3_maior_que_zero.T
```

### Visualizações

```python
fig = plt.figure(figsize =(7, 7))
plt.title('Boxplot da distribuição de defaults na amostra')
plt.boxplot(default, patch_artist=True)
```

```python
plt.title('Histograma da quantidade de ocorrência de defaults nos últimos 3 meses na amostra')
plt.hist(df_default3_maior_que_zero, bins=30, color='green')
```
